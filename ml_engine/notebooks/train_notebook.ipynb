# H&M Fashion Recommendation System Training Notebook

This notebook is designed to train the machine learning models for the H&M Personalized Fashion Recommendation System. It includes data loading, preprocessing, model training, and evaluation steps.

```python
# Import necessary libraries
import pandas as pd
from ml_engine.data_loader import load_data

# Load datasets
customers = pd.read_csv('../dataset/customers.csv')
articles = pd.read_csv('../dataset/articles.csv')
transactions = pd.read_csv('../dataset/transactions.csv')

print(customers.head())
print(articles.head())
print(transactions.head())
```

```python
# Data Preprocessing
from ml_engine.preprocess import preprocess_data

# Preprocess data
processed_data = preprocess_data(customers, articles, transactions)
print(processed_data.head())
```

```python
# Model Training
from ml_engine.train import train_models

# Train models
cf_model, text_model, image_model = train_models(processed_data)
print("Models trained successfully!")
```

```python
# Save the trained model
import joblib

joblib.dump(cf_model, '../ml_engine/model.pkl')
print("Model saved successfully!")
```

In this notebook, we have successfully loaded the data, preprocessed it, trained the machine learning models, and saved the trained model for future use.
